```python
# import the tokenizer module
# access interface functions using tk
import tokenizer as tk   
from parse import Parse
from interpreter import Interpreter
```

```python
# Tokenize code lines on 'sample.art' source file
tokendf = tk.Tokenize('sample.art')
#tokendf     # display tokenlist dataframe
```

```python
prMod = Parse('sample.art', tokendf)
prMod.Parser()
```

```python
intptr = Interpreter(tokendf)
intptr.Interpret()
intptr.Build()
```

```python
for i in range(len(intptr.tokens)):
    print("{0:20} {1}".format(
            intptr.tokens[i]["type"],
            intptr.tokens[i]["token"])
         )
```
